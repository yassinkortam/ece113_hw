```python
import numpy as np
import matplotlib.pyplot as plt

P = 100 #precision for continuous signal
FS = 10 #sampling rate

#generate n series data that approximates a continous signal over one fundamental
def signal(f):
    t0 = 1/f
    return np.arange(0, t0, t0/P), np.array([np.cos(2*np.pi*f*t) for t in np.arange(0, t0, t0/P)])

#generate n series data that samples a continous signal over one fundamental
def sample(f, fs):
    t0 = 1/f
    ts = 1/fs
    return np.arange(0, t0, ts), np.array([np.cos(2*np.pi*f*n) for n in np.arange(0, t0, ts)])

#repeat for frequencies given in homework
for f in [3, 7, 13]:
    signal_x, signal_y = signal(f)
    sample_x, sample_y = sample(f, FS)
    plt.figure()
    plt.title("F = {f}".format(f=f))
    plt.plot(signal_x, signal_y)
    plt.stem(sample_x, sample_y)
    plt.show()
```
