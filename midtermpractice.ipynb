```python
def u(n):
    if n >= 0: 
        return 1
    else:
        return 0

def mine(n):
    return (u(n)*( (1 - 0.9**(n+1))/(1-0.9) )) - ((u(n-7)*( (1 - 0.9**(n-6))/(1-0.9) )) )

def prof(n):
    return (( (1-0.9**(n+1))/(1-0.9) )*(u(n)-u(n-7))) + (( 0.9**n )*( (1 - 0.9**(-7))/(1-0.9**(-1)) )*u(n-7))
```

```python
import matplotlib.pyplot as plt
import numpy as np

mysolution = list(map(mine, np.arange(0,20,1)))
profsolution = list(map(prof, np.arange(0,20,1)))

plt.figure()
plt.plot(mysolution)
plt.plot(profsolution)
```
